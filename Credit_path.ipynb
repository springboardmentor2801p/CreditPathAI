```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
df = pd.read_csv(list(uploaded.keys())[0])

print("Original Shape:", df.shape)
df.head()
```

```python
print("Duplicate Rows:", df.duplicated().sum())
df = df.drop_duplicates()
print("After Removing Duplicates:", df.shape)
```

```python
unnecessary_cols = ['id', 'ID', 'customer_id', 'Customer_ID', 'Name', 'name']
df.drop(columns=unnecessary_cols, inplace=True, errors='ignore')
```

```python
print("Missing Values Before:\n", df.isnull().sum())

num_cols = df.select_dtypes(include=np.number).columns
cat_cols = df.select_dtypes(include='object').columns

# Numeric → Median
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# Categorical → Mode
df[cat_cols] = df[cat_cols].fillna(df[cat_cols].mode().iloc[0])

print("Missing Values After:\n", df.isnull().sum())
```

```python
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df[col] = np.clip(df[col], lower, upper)

print("Outliers Treated")
```

```python
# Example 1: Loan to Income Ratio
if 'loan_amount' in df.columns and 'annual_income' in df.columns:
    df['loan_income_ratio'] = df['loan_amount'] / df['annual_income']

# Example 2: EMI to Income Ratio
if 'monthly_income' in df.columns and 'emi' in df.columns:
    df['emi_income_ratio'] = df['emi'] / df['monthly_income']
```

```python
cat_cols = df.select_dtypes(include='object').columns

df = pd.get_dummies(df, columns=cat_cols, drop_first=True)

print("After Encoding Shape:", df.shape)
```

```python
bool_cols = df.select_dtypes(include='bool').columns
df[bool_cols] = df[bool_cols].astype(int)
```

```python
df.to_csv("Preprocessed_Loan_Dataset.csv", index=False)

from google.colab import files
files.download("Preprocessed_Loan_Dataset.csv")
```
