```
import pandas as pd
```

```
df = pd.read_csv("Loan_Default_cleaned.csv")
```

```
df.head()
```

```
df.isnull().sum()
```

```
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
```

```
df.isnull().sum()
```

```
cat_cols = ['loan_purpose', 'Neg_ammortization', 'submission_of_application']

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```
df.isnull().sum()
```

```
df[['loan_purpose', 'Neg_ammortization', 'submission_of_application']].isnull().sum()
```

```
df.isnull().sum().sum()
```

```
df.isnull().values.any()
```

```
df.duplicated().sum()
```

```
df = df.drop_duplicates()
```

```
df.duplicated().sum()
```

```
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
num_cols
```

```
df[num_cols].describe()
```

```
# @title Default title text
cat_cols = df.select_dtypes(include=['object']).columns
cat_cols
```

```
df['Status'].value_counts(dropna=False)
```

```
df = pd.get_dummies(df, drop_first=True)
```

```
X = df.drop('Status', axis=1)
y = df['Status']
```

```
df.isnull().sum()
```

```
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled_array = scaler.fit_transform(X)
```

```
X_scaled = pd.DataFrame(X_scaled_array, columns=X.columns)
X_scaled['Status'] = y.values
```

```
X_scaled.isnull().sum().sum()
```

```
X_scaled.drop('Status', axis=1).describe()
```

```
X_scaled.head()
```

```
X_scaled.isnull().sum().sort_values(ascending=False).head(10)
```

```
X_scaled.to_csv("loan_data_preprocessed.csv", index=False)
```
